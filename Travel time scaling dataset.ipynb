```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
xl_file = pd.ExcelFile("complete-data.xlsx")
```

```python
cities =pd.read_excel(xl_file)
cities.head()
```

```python
cities[cities["Urban Area"]=='Boston MA-NH-RI']
```

```python
from matplotlib import cm
fig, ax = plt.subplots()
cmap = cm.get_cmap('jet') # Colour map (there are many others)
cities["Delay by Pop"] = cities['Annual Hours of Delay'][4:-1]/cities['Population.1'][4:-1]
cities[cities["Urban Area"]=='Boston MA-NH-RI'].plot(x="Population.1",y='Delay by Pop',c='Year',kind='scatter',loglog=True,ax=ax,cmap=cmap)
ax.set_xlabel("Population")
plt.title("Boston")
plt.show()
```
